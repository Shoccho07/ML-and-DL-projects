```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
pf = pd.read_csv("/kaggle/input/shopping-annual-income/shopping_data.csv")
```

```python
pf.head()
```

```python
plt.scatter(pf['Age'],pf['Annual Income (k$)'])
```

```python
km=KMeans(n_clusters=5,n_init=10)
km
```

```python
y_predicted=km.fit_predict(pf[['Age','Annual Income (k$)']])
y_predicted
```

```python
pf['cluster']=y_predicted
pf.head
```

```python
pf1=pf[pf.cluster==0]
pf2=pf[pf.cluster==1]
pf3=pf[pf.cluster==2]
pf4=pf[pf.cluster==3]
pf5=pf[pf.cluster==4]

plt.scatter(pf1.Age,pf1['Annual Income (k$)'],color='green')
plt.scatter(pf2.Age,pf2['Annual Income (k$)'],color='red')
plt.scatter(pf3.Age,pf3['Annual Income (k$)'],color='blue')
plt.scatter(pf4.Age,pf4['Annual Income (k$)'],color='yellow')
plt.scatter(pf5.Age,pf5['Annual Income (k$)'],color='purple')

plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend()
```
